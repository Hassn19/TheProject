```python
#using sklearn to develope a Linear Regression model
import pandas as pd
import numpy as np
import sklearn as sk
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
housing = pd.read_csv('/Users/Hasan/Desktop/Housing Dataset/housing_train.csv') 
```

```python
housing.plot(kind = 'scatter', x = 'YearBuilt', y = 'SalePrice')
```

```python
# this will be the test train split, supervised learning. we create the model using only the training data.
# we then use the model to see how well it predicts the testing model
#Usually 70 percent train, 30 percent test

X_train, X_test, y_train, y_test = train_test_split(housing.YearBuilt, housing.SalePrice)
```

```python
# The test train split visulization
plt.scatter(X_train, y_train, label = 'The Training Data', color = 'r', alpha = .5)
plt.scatter(X_test, y_test, label = 'Testing Data', color = 'b', alpha = .5)

plt.legend()
plt.title("Test Train Split")
plt.show()
# Note again, the training data will be used to fit our model.
# the model will then be tested against everything in the gree
```

```python
# Here, we can begin to create the linear model
#must be 2-dimensional, hence why I put reshape(-1,1)
LR = LinearRegression()
LR.fit(X_train.values.reshape(-1,1), y_train.values)
```

```python
#Use this model to predict on the test, plot this against a scatter plot of the actual test dataset

predictions = LR.predict(X_test.values.reshape(-1,1))

# This will predict the X_test

plt.plot(X_test, predictions, label = 'Linear Regression', color = 'r')
plt.scatter(X_test, y_test, label = 'The Actual Test Dataset', color = 'b', alpha = .5)
plt.legend()
plt.show()
```

```python
# Predicting a specific a hous built in 2010

LR.predict(np.array([[2010]]))[0]
```

```python
#A sale price of a house built in 2010 would be $ 229,705
```

```python
#Score this model out of a 1.0, eh not too fantastic for this model.

LR.score(X_test.values.reshape(-1,1), y_test.values)
```
